#Inspecting a DataFrame


```python
# Finding the number of rows and columns 
diabetes_df.shape
```

```python
# First 5 rows in a DataFrame
diabetes_df.head()
```

```python
# Last 5 rows of a DataFrame
diabetes_df.tail()
```

```python
# Finding number of missing values
diabetes_df.isnull().sum()
```

Statistical Measures

```python
# Count number of values
diabetes_df.count()
```

```python
# Calculating the mean value for each column
diabetes_df.mean()
```

```python
# Standard Deviation based on column
diabetes_df.std()
```

```python
# Maximum value for each column 
diabetes_df.max()
```

```python
# Minimum value for each column 
diabetes_df.min()
```

```python
# All the statistical measures about the DataFrame
diabetes_df.describe()
```

Manpuating the DataFrame

```python
# Adding a column to a DataFrame
diabetes_df['BlurredVision'] = 1
```

```python
diabetes_df.head()
```

```python
# Removing a row 
diabetes_df.drop(index = 0, axis = 0)
```

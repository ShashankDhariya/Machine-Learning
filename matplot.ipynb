#Matplot Library 
    Useful for making Plots

```python
import matplotlib.pyplot as plt
```

```python
# Import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# Sine wave ploting
plt.plot(x, y)
plt.show()
```

```python
# Cosine wave ploting 
plt.plot(x, z)
plt.show()
```

```python
# Adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine Wave')
plt.show()
```

```python
# Ploting a parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# Ploting with symbol & color
plt.plot(x, y, 'r+')
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'b--')
plt.show()
```

```python
x = np.linspace(-20, 24, 50)
plt.plot(x, x**2, 'g*')
plt.show()
```

Bar Plot 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Hindi', 'Malyalam', 'French', 'German']
people = [150, 290, 50, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGE')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie Chart 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Hindi', 'Malyalam', 'French', 'German']
people = [150, 290, 50, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 40)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='b')
ax.scatter(x,z,color='g')
plt.show()
```

3D Scatter Plot

```python
fig3 =  plt.figure()
ax = plt.axes(projection = '3d')
x = 20* np.random.random(100)
y = np.sin(x)
z = np.sin(x)
ax.scatter(x,y,z,cmap='Blues')
plt.show()
```
